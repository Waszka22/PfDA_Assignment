 The seed() method is used to initialize the random number generator. The random number generator needs a number to start with, to be able to generate a random number. By default the random number generator uses the current system time. [1].



```python
# the seed() method is used to initialize the random number generator
np.random.seed(42)

# generate random data for each numerical variable
for column in df.select_dtypes(include=[np.number]).columns:
    mean, std = df[column].mean(), df[column].std()
    random_distribution = np.random.normal(loc=mean, scale=std, size=len(df))
    df[column] = df[column] + random_distribution

# Display the updated DataFrame
print(df.head())
```

## Data Visualization

## Type of Data:
* Numerical Data 
  + Discrete: values vary by finite steps 
  + Continous: can take any numerical value
* Categorlcal Data 
  + Nominal: categories have no particular order
  + Ordinal: there is a natural ordering to the categories
  


### Numerical Variables:
+ Age: numeric (continuous), range between 27 to 59 years 
+ Sleep Duration: numeric (continuous) bettwen 5.8 and 8.5 hours
+ Quality of Sleep: numeric (discrete)
+ Physical Activity Level: numeric (continuous)
+ Stress Level: numeric (discrete)
+ Blood Pressure: numeric (continuous)
+ Heart Rate: numeric (continuous) range between 65 to 86 bpm 
+ Daily Steps: numeric (continuous)


```python
# numerical variables
numerical_vars = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']
for var in numerical_vars:
    plt.figure(figsize=(8, 5))
    plt.hist(df[var], bins='auto', color='plum', edgecolor='black')
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
   
    plt.tight_layout()
    plt.show()
```

### Categorical Variables:
+ Gender: Male (nominal)
+ Occupation: Software Engineer, Doctor, Sales Representative, Teacher (nominal)
+ BMI Category: Underweight, Normal, Overweight, Obese (nominal)
+ Sleep Disorder: None, Sleep Apnea, Insomnia (nominal)

#### Bar Chart of Gender



```python
# adding bar chart of gender
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts, color=['darkturquoise', 'hotpink'])

# add frequency numbers 
for i, count in enumerate(gender_counts):
    plt.text(i, count + 1, str(count), ha='center', va='bottom')

plt.title('Bar Chart of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
```

#### Histagram of BIM Category

```python
# An "interface" to matplotlib.axes.Axes.hist() method
n, bins, patches = plt.hist(x=df['BMI Category'], bins='auto', color='#98F5FF')
# add grid                    
plt.grid(axis='y', alpha=0.75)
# add a lable for plot
plt.xlabel('BIM Category')
plt.ylabel('Count')
#add grid
plt.grid(True)
# add a title for plot
plt.title('Histagram of BMI Category')
```

```python
# Creating dataset
sleep_count = df['Sleep Disorder'].value_counts()
custom_colors = ['#00FFFF', '#EE1289', '#99FF99']
plt.figure(figsize=(8, 8))
plt.pie(sleep_count, autopct='%1.1f%%', startangle=135, colors=custom_colors)
# Adding title 
plt.title("Sleep Disorder Distribution", loc='left') 
# Adding legend
plt.legend(labels=sleep_count.index, title="Sleep Disorder Type", loc="upper right")
```

#### Relationship beetwen Stress Level vs. Quality of Sleep

To visualize the relationship between two variables, we use the 'regplot' function from the seaborn library. 'regplot' creates a scatter plot with a linear regression line. The 'scatter_kws parameter' is used to adjust the transparency of points. 


```python
# scatter plot with linear regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Stress Level', y='Quality of Sleep', data=df, color='teal', scatter_kws={'alpha':0.7})
# add a title for plot
plt.title('Stress Level vs. Quality of Sleep')
# add a lable for plo
plt.xlabel('Sleep Duration')
plt.ylabel('Quality of Sleep')
#add grid
plt.grid(True)
plt.show()
```

 + The scatter plot shows lower stress levels associated with higher sleep quality.
 + A negative correlation, suggesting that as stress levels increase, sleep quality tends to deteriorate.

#### Relationship beetwen  Sleep Duration vs. Quality of Sleep

```python
# scatter plot with linear regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Sleep Duration', y='Quality of Sleep', data=df, color='plum', scatter_kws={'alpha':0.7})
plt.title('Sleep Duration vs. Quality of Sleep')
# add a lable for plo
plt.xlabel('Sleep Duration')
plt.ylabel('Quality of Sleep')
#add grid
plt.grid(True)
plt.show()
```

+ There is a positive correlation suggesting that as sleep duration increases, sleep quality improves.
